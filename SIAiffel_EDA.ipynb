## 1. Buildings 

```python
B_train_json = os.getenv('HOME')+'/SIAIFFEL_Dataset/buildings/Training/라벨링데이터'

B_train_json = sorted(glob.glob(B_train_json + "/*json"))
len(B_train_json)
```

__EDA를 위한 csv변환__

1. Site.csv
2. Road.csv
3. Building.csv
4. BnR.csv

```python
B_train_json
```

```python
B_train_list = []

for i in B_train_json:
    B_json_feature = json.load(open(i))
    
    for j in B_json_feature['features']:
        
        # Geometry 
        geo_coord = j['geometry']['coordinates']
        geo_type = j['geometry']['type']
        
        # Properties
        type_id = j['properties']['type_id']
        type_name = j['properties']['type_name']
        object_imcoords = j['properties']['object_imcoords']
        building_imcoords = j['properties']['building_imcoords']
        road_imcoords = j['properties']['road_imcoords']
        image_id = j['properties']['image_id']
        ingest_time = j['properties']['ingest_time']
        
        # append B_train_list
        B_train_list.append([image_id, type_name, type_id, geo_coord, geo_type,
                           object_imcoords, building_imcoords, road_imcoords, ingest_time])
```

```python
B_train = pd.DataFrame(B_train_list,
                        columns = ['image_id','type_name','type_id','geo_coord','geo_type',
                                   'object_imcoords','building_imcoords','road_imcoords','ingest_time'])
```
